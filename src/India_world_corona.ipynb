```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('datasets_mind7/COVID-19_cases_timeseries.csv')
df2 = pd.read_csv('datasets_mind7/country_wise_death_timeseries.csv')
df3 = pd.read_csv('datasets_mind7/recovered_cases.csv')
```

```python
ls = df1.columns
dt = 'date'
ls = [str(i)+"_case" for i in ls]
ls[0] = dt
df1.columns = ls
```

```python
ls = df2.columns
dt = 'date'
ls = [str(i)+"_death" for i in ls]
ls[0] = dt
df2.columns = ls
```

```python
ls = df3.columns
dt = 'date'
ls = [str(i)+"_recov" for i in ls]
ls[0] = dt
df3.columns = ls
```

```python
df = pd.merge(df1,df2,on='date')
```

```python
df = pd.merge(df,df3,on='date')
```

```python
India = df[['date','India_case','India_death', 'India_recov']]
```

```python
#India = India.iloc[:-1,:]
ax = np.array(['3/23/20',470,9.0,40]).reshape(1,4)
ls = pd.Series(ax)
# d1f = pd.DataFrame(ls)
# d1f.shape
```

```python
X = np.arange(1,62).reshape(-1,1)
y = India[['India_case']].values
len(y)
```

```python
fig,ax = plt.subplots()
ax2 = ax.twinx()
ax.plot(India['date'], India['India_case'], marker='s')
ax.set_xticklabels(India['date'],rotation=90)
fig.set_size_inches(18,10)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LassoCV
from sklearn.pipeline import make_pipeline

lasso_eps = 0.0001
lasso_nalpha=20
lasso_iter=5000

degree_min = 1
degree_max = 20

for degree in range(degree_min,degree_max+1):
    model = make_pipeline(PolynomialFeatures(degree, interaction_only=False), LassoCV(eps=lasso_eps,n_alphas=lasso_nalpha,max_iter=lasso_iter,
normalize=True,cv=5))
    model.fit(X,y)
    test_pred = np.array(model.predict(X))
    #RMSE=np.sqrt(np.sum(np.square(test_pred - y_test)))
    test_score = model.score(X,y)
```

```python
y_pred = model.predict(X)
y_pred = [int(i) for i in y_pred]
```

```python
y = [int(i) for i in y]
```

```python
fig,ax = plt.subplots()
ax.plot(India['date'],y_pred, marker='s')
ax.plot(India['date'],y, marker='s')
ax.set_xticklabels(India['date'],rotation=90)
ax.set_ylabel("Number of cases")
ax.set_xlabel("Date")
ax.tick_params('y', colors='blue')
ax.legend(['predicted','actual'])

ax.annotate('Lockdown starts here', xy = ('3/22/20',2), xytext = ('2/22/20', 360), arrowprops={'arrowstyle':'->','color':'red'})

ax2 = ax.twinx()
ax2.plot(India['date'],India['India_death'], marker='v')
ax2.plot(India['date'],India['India_recov'], marker='o', color='black', linestyle='--')
ax2.set_ylabel("cases")
ax2.legend(['death','recovered'],loc=1)
ax2.tick_params('y', colors='red')
fig.set_size_inches(18,10)
plt.title("Corona cases In India [ only case is predicted here ]")
plt.show()
```
